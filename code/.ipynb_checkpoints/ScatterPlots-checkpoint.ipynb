```python
# Basic scatter plots for summary data 
x_stars = df['stars'].tolist()
y_review_count = df['review_count'].tolist()
y_median = df['Median'].tolist()
y_pop = df['Pop'].tolist()
y_checkins = df['checkins'].tolist()
```

```python
plt.scatter(x_stars,y_review_count, color = 'gold', edgecolor = 'blue')
plt.xlabel('Stars')
plt.ylabel('Review Count')
plt.title("Stars compared to Review Count")
plt.show()
```

```python
plt.scatter(x_stars,y_checkins, color = 'gold',  edgecolor = 'blue')
plt.xlabel('Stars')
plt.ylabel('checkins')
plt.title("Stars compared to Checkins")
plt.show()
```

```python
plt.scatter(y_review_count,y_checkins, color = 'gold',  edgecolor = 'blue')
plt.xlabel('Review Count')
plt.ylabel('Checkins')
plt.title("Review Count Compared To Checkins ")
plt.show()
```

```python
plt.scatter(x_stars,y_pop, color = 'gold',  edgecolor = 'blue')
plt.xlabel('Stars')
plt.ylabel('Population')
plt.title("Stars compared to Population")
plt.show()
```

```python
plt.scatter(x_stars,y_median, color = 'gold',  edgecolor = 'blue')
plt.xlabel('Stars')
plt.ylabel('Median Income')
plt.title("Stars compared to Median Income")
plt.show()
```
