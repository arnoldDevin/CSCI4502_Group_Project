```python
# http://pbpython.com/market-basket-analysis.html
# https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/
#mlxtend is a python library with tools used to data science and machine learning
#If using jupyter through anaconda, download here: https://anaconda.org/conda-forge/mlxtend
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#Use clean_test.csv file
df = pd.read_csv('yelp-dataset/clean_test.csv', low_memory = False, index_col=0)
```

```python
#Replace values with more than 2 unique members with integers
df['Alcohol'].replace(to_replace = 'full_bar', value =2 , inplace = True )
df['Alcohol'].replace(to_replace = 'beer_and_wine', value =1 , inplace = True )

df['BYOB'].replace(to_replace = 'yes_free', value =1 , inplace = True )
df['BYOB'].replace(to_replace = 'no', value =0 , inplace = True )
df['BYOB'].replace(to_replace = 'yes_corkage', value =1 , inplace = True )

df['AgesAllowed'].replace(to_replace = 'allages', value =1 , inplace = True )
df['AgesAllowed'].replace(to_replace = '18', value =0 , inplace = True )
df['AgesAllowed'].replace(to_replace = '21', value =0 , inplace = True )


```

```python
df = df.set_index('business_id')
```

```python
df.drop('Alcohol', inplace=True, axis=1)
df.drop('BYOB', inplace=True, axis=1)
df.drop('AgesAllowed', inplace=True, axis=1)
```

```python
#First 5 rows of dataframe
df.head()
#df['GoodForKids'].unique()
```

```python
#Minimum support low simply to see the most amount of Lift correlations
frequent_itemsets = apriori(df, min_support=0.001, use_colnames=True)
```

```python
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', None)  
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=100)
rules.head(1000)

#115 Combinations with Lift >= 1
#35 Combinations with Lift >= 50
#17 Combinations with Lift >= 100
```
