```python
# http://pbpython.com/market-basket-analysis.html
# https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('clean_test.csv', low_memory = False, index_col=0)
```

```python
df['Alcohol'].replace(to_replace = 'full_bar', value =2 , inplace = True )
df['Alcohol'].replace(to_replace = 'beer_and_wine', value =1 , inplace = True )

df['BYOB'].replace(to_replace = 'yes_free', value =1 , inplace = True )
df['BYOB'].replace(to_replace = 'no', value =0 , inplace = True )
df['BYOB'].replace(to_replace = 'yes_corkage', value =1 , inplace = True )

df['AgesAllowed'].replace(to_replace = 'allages', value =1 , inplace = True )
df['AgesAllowed'].replace(to_replace = '18', value =0 , inplace = True )
df['AgesAllowed'].replace(to_replace = '21', value =0 , inplace = True )


```

```python
df = df.set_index('business_id')
```

```python
df.drop('Alcohol', inplace=True, axis=1)
df.drop('BYOB', inplace=True, axis=1)
df.drop('AgesAllowed', inplace=True, axis=1)
```

```python
df.head()
```

```python
frequent_itemsets = apriori(df, min_support=0.007, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules.head(100)
```
